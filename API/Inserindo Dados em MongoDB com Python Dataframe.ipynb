```python
import requests
import pandas as pd
from pymongo import MongoClient

url = "https://servicodados.ibge.gov.br/api/v1/localidades/distritos"
response = requests.get(url)
data = response.json()
df = pd.json_normalize(data) 

colunas = ['nome',
           'id',
           'municipio.microrregiao.mesorregiao.UF.regiao.sigla',
           'municipio.regiao-imediata.regiao-intermediaria.UF.regiao.nome'
           ]

df = df[colunas].rename(columns={
    'nome':'Distrito',
    'municipio.microrregiao.mesorregiao.UF.regiao.sigla':'UF',
    'municipio.regiao-imediata.regiao-intermediaria.UF.regiao.nome':'regiao' 
        
    })

#df.head()
df_dicionario = df.to_dict(orient='records')

#Mapeamento de destino
client = MongoClient('mongodb+srv://eu:teste@cluster01.8bpifsh.mongodb.net/')
db = client['IBGE']
collection= db['distritos_dataframe']

#delete antes da carga
collection.delete_many({})

#enviando dados para MongoDB
collection.insert_many(df_dicionario)
```
